```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('googleplaystore.csv')
data.head()
```

```python
# using function to get details of the data easily
def head(x):
    return data.head(x)
def tail(x):
    return data.tail(x)
def info():
    return data.info()
def columns():
    return list(data.columns)
def isnull():
    return data.isnull().sum()
def shape():
    return data.shape
```

```python
# null values
isnull()
```

```python
shape()
```

```python
# Droping Null values
data.dropna(inplace=True)
shape()
```

```python
head(10)
```

```python
isnull()
```

```python
data['Installs'].unique()
```

```python
# Processing Variables of Installs Columns
def install(x):
    i = float(x.replace(',','')[:-1])
    return float(i)
```

```python
data['Installs'] = data['Installs'].apply(install)
head(10)
```

```python
# Processing Variables of Size Columns
data['Size'].unique()
def size(x):
    if 'M' in x:
        i = float(x[:-1])
        return i * 1000
    elif 'k'in x:
        i = float(x[:-1])
        return i
    else:
        return float(np.nan)
```

```python
data['Size_in_kb'] = data['Size'].apply(size)
head(10)
```

```python
info()
```

```python
# changing variables of review column to float
data['Reviews'] = data['Reviews'].astype(float)
```

```python
# '''Removing rows with reviews more than installs as an app cannot be reviewed without installation and experience'''
list(data[(data['Reviews']>data['Installs'])].index)
```

```python
for i in list(data[(data['Reviews']>data['Installs'])].index):
    data.drop(i,inplace=True)
```

```python
list(data[(data['Reviews']>data['Installs'])].index)
```

```python
head(5)
```

```python
# Processing Variables of Price
data['Price'].unique()
```

```python
data['Price'] = data['Price'].apply(lambda x:float(x.replace('$','')))
```

```python
data['Price'].unique()
```

```python
info()
```

```python
head(5)
```

```python
# Applying datatime function to data variables
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
```

```python
head(10)
```

```python
# Checking for Duplicate variable
data[data.duplicated(subset='App')]
```

```python
# Droping Duplicates
data.drop_duplicates(subset='App',inplace=True)
```

```python
data[data.duplicated(subset='App')]
```

## Exploratory Data Analysis

```python
# Apps with highest downloads
Highest_downloaded_Apps = data[data['Installs']==data['Installs'].max()]
print(Highest_downloaded_Apps[['App','Installs']].to_string(index=False))
```

```python
head(10)
```

```python
# Biggest Apps in size KB 
Biggest_Apps = data[data['Size_in_kb']==data['Size_in_kb'].max()]
print(Biggest_Apps[['App','Size','Size_in_kb']].to_string(index=False))
```

```python
# Free Apps with most reviews 
most_reviewed_apps = data.sort_values(by='Reviews',ascending=False)
most_reviewed_apps[['App','Reviews','Type']][:10]
```

```python
# Paid Apps with most reviews 
data[data['Type']=='Paid'].sort_values(by='Reviews',ascending=False)[:10][['App','Reviews','Type']]
```

```python
# Paid apps with top rating
data[(data['Type']=='Paid') & (data['Rating']==5.0)][['App','Type','Rating']]
```

```python
# average rating of apps as per their category
data.groupby('Category')['Rating'].mean().sort_values(ascending=False)
```

```python
# Number of Apps for specific kind of people
data.groupby('Content Rating').size()
```

```python
for x,y in zip(data.groupby('Content Rating').size(),data.groupby('Content Rating').size().keys()):
    if x > 1:
        print('There are total',x,'Apps for',y.upper())
    else:
        print('There is',x,'App for',y.upper())
```

```python
# Number of apps as per Category
data.groupby('Category').size()
for x,y in zip(data.groupby('Category').size().keys(),data.groupby('Category').size()):
    if y > 1:
        print('There are total',y,'Apps for',x.upper())
    else:
        print('There are total',y,'App for',x.upper())
```

```python
# Top 5 Category with most apps
data['Category'].value_counts().sort_values(ascending=False)[:5].plot(kind='bar')
plt.xticks(rotation=0)
plt.title('Top 5 Category with most apps')
plt.show()
print(data.groupby('Category').size().sort_values(ascending=False)[:5].to_string())
```

```python
# Costliest Apps
data.sort_values(by=['Price'],ascending=False)[:10][['Category','App','Price']]
```

```python
data.sort_values(by=['Price'],ascending=False)[data['Price']<399.99][:10][['Category','App','Price']]
```
