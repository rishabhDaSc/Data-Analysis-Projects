```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
# number of people survived 
plt.figure(figsize=(5,2))
df['alive'].value_counts().plot(kind='bar',color=['black','red'])
plt.title('Number of people survived')
plt.xticks(rotation=0)
plt.show()
print(df['alive'].value_counts().to_string())
```

```python
# average fare price of class
plt.figure(figsize=(8,4))
df.groupby('class')['fare'].mean().plot(kind='bar',color=['r','b','g'])
plt.title('Average fare price of class')
plt.xticks(rotation=0)
plt.xlabel('')
plt.show()
print(df.groupby('class')['fare'].mean().to_string())
```

```python
# Age distribution of passengers 
plt.figure(figsize=(8,4))
plt.hist(df['age'],bins=80,color='brown')
plt.title('Age distribution of passengers')
plt.xlim(0,80)
plt.show()
print('\tMost of the passengers on titanic were of age between 10 and 40')
```

```python
# bar graph showing number of passengers from different towns
plt.figure(figsize=(8,3))
plt.bar(df['embark_town'].value_counts().keys(),df['embark_town'].value_counts(),color='orange')
plt.title('bar graph showing number of passengers from different towns',size=10)
plt.xticks(rotation=0)
plt.show()
print(df['embark_town'].value_counts().to_string())
```

```python
# passengers of which town is in which class
plt.figure(figsize=(8,3))
df.groupby('embark_town')['class'].value_counts().plot(kind='barh')
plt.ylabel('')
plt.title('Passengers from Town and Class',size=8)
plt.yticks(size=8)
plt.show()
print(df.groupby('embark_town')['class'].value_counts().to_string())
```

```python
# Number of passengers travelling along
plt.figure(figsize=(8,3))
df['alone'] = df['alone'].map({True:'Travelling Alone',False:'Not Alone'})
df['alone'].value_counts().plot(kind='bar',color=['b','brown'])
plt.ylabel('')
plt.title('Number of passengers travelling along',size=8)
plt.xticks(size=8,rotation=0)
plt.show()
print(df['alone'].value_counts().to_string())
```

```python
# number of man,woman and child from towns
plt.figure(figsize=(8,3))
df.groupby('embark_town')['who'].value_counts().plot(kind='barh',color='r')
plt.ylabel('')
plt.title('Number of Man and Woman from Towns',size=8)
plt.yticks(size=8)
plt.show()
print(df.groupby('embark_town')['who'].value_counts().to_string())
```

```python
# pie chart showing percent of man and woman traveling alone and not alone
plt.figure(figsize=(10,5))
plt.pie(df.groupby('alone')['who'].value_counts(),normalize=True,labels=df.groupby('alone')['who'].value_counts().keys(),autopct='%.2f%%')
plt.ylabel('')
plt.title('Number of Man and Woman from Towns',size=8)
plt.yticks(size=8)
plt.show()
```
